```python
# ============================================================
# TASK 3: CUSTOMER SUPPORT CHATBOT
# Using TF-IDF + Cosine Similarity + Gradio
# (Self-contained, no file upload needed)
# ============================================================

# 1. Install Libraries
!pip install gradio --quiet

# 2. Import Libraries
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

import gradio as gr

# ============================================================
# 3. CREATE A SMALL DATASET INSIDE COLAB (NO UPLOAD NEEDED)
# ============================================================

data = {
    "question": [
        "How can I reset my password?",
        "What is your refund policy?",
        "How do I track my order?",
        "What payment methods do you accept?",
        "How can I contact customer support?",
        "How do I change my delivery address?",
        "Do you offer international shipping?",
        "How can I cancel my order?",
        "What should I do if I received a damaged product?",
        "How do I apply a discount code?"
    ],
    "answer": [
        "You can reset your password by clicking on the 'Forgot Password' link on the login page and following the instructions.",
        "We offer a 7-day refund policy from the date of delivery. Please check our Refund Policy page for more details.",
        "You can track your order using the tracking link sent to your registered email or by visiting the 'My Orders' section.",
        "We accept credit cards, debit cards, net banking, UPI, and popular wallet payments.",
        "You can contact customer support via email, live chat, or our toll-free helpline number listed on the Contact Us page.",
        "You can change your delivery address from the 'My Orders' section before the order is shipped.",
        "Yes, we offer international shipping to selected countries. Please check the Shipping Policy page for the full list.",
        "You can cancel your order from the 'My Orders' section before it is shipped. Once shipped, cancellation may not be possible.",
        "If you received a damaged product, please raise a complaint through the Help Center within 48 hours with photos of the product.",
        "You can apply a discount code on the checkout page in the 'Apply Coupon' section before making payment."
    ]
}

df = pd.DataFrame(data)

# Save to CSV (for reference, not mandatory to use later)
df.to_csv("customer_support_small.csv", index=False)

print("âœ… Sample dataset created with columns: ", df.columns.tolist())
print(df.head())

# ============================================================
# 4. PREPARE DATA FOR TF-IDF
# ============================================================

question_col = "question"
answer_col = "answer"

# Select only needed columns and drop missing values
df = df[[question_col, answer_col]].dropna().reset_index(drop=True)

# List of all questions
questions = df[question_col].astype(str).tolist()

# Create TF-IDF vectorizer
vectorizer = TfidfVectorizer(stop_words='english')

# Fit and transform questions
tfidf_matrix = vectorizer.fit_transform(questions)

print("\nâœ… TF-IDF matrix created successfully!")
print("   Number of FAQs:", len(questions))

# ============================================================
# 5. CHATBOT FUNCTION
# ============================================================

def get_answer(user_query):
    """
    Takes the user query and returns the best matching answer
    from the dataset using cosine similarity.
    """
    user_query = user_query.strip()

    if user_query == "":
        return "Please type your question ðŸ™‚"

    # Transform user query into TF-IDF vector
    user_tfidf = vectorizer.transform([user_query])

    # Compute cosine similarity with all FAQ questions
    cosine_similarities = cosine_similarity(user_tfidf, tfidf_matrix).flatten()

    # Get index of best match
    best_idx = cosine_similarities.argmax()
    best_score = cosine_similarities[best_idx]

    # Set a threshold for "I don't know" type responses
    threshold = 0.2

    if best_score < threshold:
        return (
            "Sorry, I am not very sure about this query. ðŸ˜•\n"
            "Please try rephrasing your question or contact human support."
        )

    # Get best matched question and answer
    best_question = df.loc[best_idx, question_col]
    best_answer = df.loc[best_idx, answer_col]

    # Nicely formatted output
    response = (
        f"ðŸ’¬ **Customer Support Bot**\n\n"
        f"**Answer:**\n{best_answer}\n\n"
        f"_(Closest FAQ: \"{best_question}\")_"
    )
    return response

# Quick test in console
print("\nðŸ” Sample Test:")
print(get_answer("How can I reset my password?"))

# ============================================================
# 6. GRADIO CHATBOT UI
# ============================================================

demo = gr.Interface(
    fn=get_answer,
    inputs=gr.Textbox(lines=2, placeholder="Type your customer query here..."),
    outputs="markdown",
    title="Customer Support Chatbot (TF-IDF + Cosine Similarity)",
    description="Ask any customer support question. The bot finds the closest FAQ answer from the built-in dataset."
)

demo.launch()
```
